# Performing analysis on the dataset.

### Overall effect of listening to music. Is there improvement, no effect or worsening.

```python
# Code obtained with help from ChatGPT
# Select the columns
mental_health_columns = ['Anxiety', 'Depression', 'Insomnia', 'OCD', 'MusicEffects']

# Create a new DataFrame with the selected columns
mental_health_df = Music_df[mental_health_columns]

# Drop rows with missing values in 'MusicEffects'
mental_health_df = mental_health_df.dropna(subset=['MusicEffects'])

# Create a pie chart
plt.figure(figsize=(10, 8))

# Assuming 'MusicEffects' is a categorical column with different values
sns.countplot(data=mental_health_df, x='MusicEffects', palette='Pastel1')

plt.title('The Effects of Music on Mental Health')
plt.xlabel('Music Effects')
plt.ylabel('Number of People')

plt.show()
```

### Looking to calculate the average hours a day, for each genre

```python
# Code obtained with help from ChatGPT
# Filtering out rows with missing values in 'favGenre' or 'HoursPerDay'
filtered_df = Music_df.dropna(subset=['favGenre', 'HoursPerDay'])

# Group by 'favGenre' to calculate avg. hours/day of listenting to partciular genres.
average_hours_per_day_by_genre = filtered_df.groupby('favGenre')['HoursPerDay'].mean()

# Displays results
print(average_hours_per_day_by_genre)

# Creating a viz in the form of a pie chart.
plt.figure(figsize=(10, 8))
plt.pie(average_hours_per_day_by_genre, labels=average_hours_per_day_by_genre.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Average Hours Per Day Listened to Each Genre')
plt.show()
```

### Looking at average hours a day of listening to music, by method (meaning how they got the music)

```python
# Code obtained with help from ChatGPT
# Filtering out the rows that have missing values in columns 'HoursPerDay' or 'streamingService'
filtered_df = Music_df.dropna(subset=['HoursPerDay', 'streamingService'])

# Creating a viz, bar chart.
plt.figure(figsize=(12, 8))
sns.barplot(x='streamingService', y='HoursPerDay', data=filtered_df, ci=None, palette='viridis')

plt.title('Average Hours Per Day by Method')
plt.xlabel('Method of Listening to Music')
plt.ylabel('Average Hours Per Day')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels to make the easier to read.

plt.show()
```

### Looking at average hours per day of music listened to, by efficacy (i.e., did it help or not)

```python
# Code obtained with help from ChatGPT
# Filter out rows with missing values in 'MusicEffects' or 'HoursPerDay'
filtered_df = Music_df.dropna(subset=['MusicEffects', 'HoursPerDay'])

# Create a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='MusicEffects', y='HoursPerDay', data=filtered_df, ci=None, palette='muted')

plt.title('Average Hours of Music Listened to Per Day by Efficacy')
plt.xlabel('Music Effects')
plt.ylabel('Average Hours Per Day')

plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

plt.show()
```

### Looking at the conditions, specifically, how many people marked that they were experiencing the condition. 

```python
# Code obtained with help from ChatGPT
# Seleting the columns
mental_health_columns = ['Depression', 'Anxiety', 'OCD', 'Insomnia']

# Counting the number of non-zero values for each column, meaning the participant had the condition.
non_zero_counts = Music_df[mental_health_columns].astype(bool).sum()

# Displaying the counts
print("Number of people with a number other than 0 in each mental health column:")
print(non_zero_counts)
```

### Looking to create a viz for the above information.

```python
# Code obtained with help from ChatGPT
# Selecting the columns
mental_health_columns = ['Depression', 'Anxiety', 'OCD', 'Insomnia']

# Count the number of non-zero values for each column
non_zero_counts = Music_df[mental_health_columns].astype(bool).sum()

# Create a bar chart
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=non_zero_counts.index, y=non_zero_counts.values, palette="viridis")
ax.set(title='Number of People who Ranked a Problem with a 1 or Greater',
       xlabel='Specific Mental Health Problems', ylabel='Number of People')

# Display the counts on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

### looking at those participants who ranked a more severe score for Anxiety. I chose 7 or greater. 

```python
# Code obtained with help from ChatGPT
# Selecting the columns
mental_health_columns = ['Depression', 'Anxiety', 'OCD', 'Insomnia']

# Count the number of particpants with a score of 7 or greater in each column, considering this score and above more severe.
severe_counts = (Music_df[mental_health_columns] >= 7).sum()

# Creating a bar chart for visualization.
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=severe_counts.index, y=severe_counts.values, palette="mako")
ax.set(title='Number of People who Ranked a Problem with a 7 or Greater',
       xlabel='Specific Mental Health Problems', ylabel='Number of People')

# Showing counts on top of each bar on graph.
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

### Looking to break out some pieces of the pie to hightlight them.

```python
# Code obtained with help from ChatGPT
genres = ['Classical', 'Country', 'EDM', 'Folk', 'Gospel', 'Hip hop', 'Jazz', 'K pop',
          'Latin', 'Lofi', 'Metal', 'Pop', 'R&B', 'Rap', 'Rock', 'Video game music']

frequencies = [2.882075, 3.42, 4.594595, 3.236667, 2.375, 3.928571, 5.425, 4.0,
                6.666667, 4.2, 3.596591, 2.907895, 3.471429, 5.318182, 3.634043, 2.914773]

# Identifying indices for 'Rap', 'Jazz', and 'Latin'
rap_index = genres.index('Rap')
jazz_index = genres.index('Jazz')
latin_index = genres.index('Latin')

# Breaking out 'Rap', 'Jazz', and 'Latin' from the rest
explode = [0] * len(genres)
explode[rap_index] = 0.1  # Separate 'Rap' slice
explode[jazz_index] = 0.1  # Separate 'Jazz' slice
explode[latin_index] = 0.1  # Separate 'Latin' slice

# Creating a pie chart for viz
plt.figure(figsize=(10, 10))
plt.pie(frequencies, labels=genres, autopct='%1.1f%%', startangle=140, explode=explode)

plt.title('Distribution of Music Genres Frequencies')

plt.show()
```
